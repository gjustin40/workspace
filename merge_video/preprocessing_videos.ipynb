```python
import cv2
import glob
import os
import numpy as np
from moviepy.editor import VideoFileClip, concatenate_videoclips

import pandas as pd

from tqdm import tqdm
import time
```

```python
weather_list = ['foggy', 'sunny', 'rainy']
base_url = 'T:/public_data/video_classification/used_video/'
foggy_list = glob.glob(base_url + 'foggy/*')
sunny_list = glob.glob(base_url + 'sunny/*')
rainy_list = glob.glob(base_url + 'rainy/*')

print('foggy num', len(foggy_list))
print('sunny num', len(sunny_list))
print('rainy num', len(rainy_list))
```

```python
filenames = foggy_list
presets = list(map(lambda x: x[-7:-4], filenames))
df_foggy = pd.DataFrame({'filename': filenames, 'preset': presets})
df_foggy['weather'] = 'foggy'

df_foggy
```

```python
preset_list = ['p01', 'p02', 'p03', 'p04', 'p05', 'p06', 'p07', 'p08']
for preset in preset_list:
    df_preset = df_foggy.loc[df_foggy['preset'] == preset]
    filenames = sorted(df_preset.filename.tolist(), key=lambda x: os.path.basename(x).split('_')[1:3])
    
    clip_list = []
    for filename in tqdm(filenames):
        clip_list.append(VideoFileClip(filename).subclip(2, -2))
        
    final_clip = concatenate_videoclips(clip_list)
    final_clip.write_videofile(f'T:/public_data/video_classification/used_video/foggy_merge/{preset}_foggy_merge.mp4')
```
